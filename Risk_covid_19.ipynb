```python
serious=[]
```

```python
i=0
while i<len(df_new):
    #print(i)
    if df_new.iloc[i]['survival']=='N':
        #rint("A")
        serious.append(1)
    else:
        #print("C")
        if df_new.iloc[i]['intubated']=='Y' or df_new.iloc[i]['intubation_present']=='Y' or df_new.iloc[i]['went_icu']=='Y' or df_new.iloc[i]['in_icu']=='Y':
            #print("B")
            serious.append(1)
        else:
            serious.append(0)
            
    #print(len(serious))
    i+=1
        

        
    
```

```python
import tensorflow as tf
```

```python
model=tf.keras.models.load_model('Covid_19_detection.h5')
```

```python
model
```

```python
import cv2
import numpy as np
def prediction(model,df):
    i=0
    pred=[]
    while i<len(df):
        
        file_ext=fldr+'/'+files[1]+'/'
        name=df.iloc[i]['filename']
        file=file_ext+name
        image=cv2.imread(file)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (224, 224))
        images_final=np.array(image)/255
        pred_1=model.predict(np.array([images_final]))
        #print(pred_1[0][0])
        pred.append(pred_1[0][0])
        
        i+=1
    return pred
        
        
        
        
        
```

```python
pred=prediction(model,df_temp)
```

```python
len(pred)
```

```python
i=0
gender=[]
while i<len(df):
    if df.iloc[i]['sex']=='M':
        gender.append(1)
    else:
        gender.append(0)
    i+=1
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X=df[['gender','age','pred']].values
Y=df['serious'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

```python
model_lr=LogisticRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=model_lr.predict(X_test)
```

```python
acc=accuracy_score(y_pred,y_test)
```

```python
acc
```

```python
model_gb=GradientBoostingClassifier()
```

```python
model_gb.fit(X_train,y_train)
```

```python
y_pred=model_gb.predict(X_test)
```

```python
acc=accuracy_score(y_pred,y_test)
```

```python
acc
```

```python
import xgboost as xgb
```
